```python
from tensorflow.compat.v1 import ConfigProto, InteractiveSession
config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.3
session = InteractiveSession(config=config)
```

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
```

```python
x_data = [[0, 0],
         [0, 1],
         [1, 0],
         [1, 1]]
y_data = [[0],
         [1],
         [1],
         [0]]
```

```python
x_data = np.array(x_data)
y_data = np.array(y_data)
```

```python
x_data.shape
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Dense(4, activation='tanh', input_shape=(2,)))
model.add(keras.layers.Dense(2, activation='tanh'))
model.add(keras.layers.Dense(1, activation='tanh'))
```

```python
model.compile(optimizer='SGD', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x_data, y_data, batch_size=4, epochs=2000)
```

```python
model.predict(x_data)
```

```python
np.round(model.predict(x_data))
```

# Mnist 를 학습시켜 보자.

```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
X_train.shape
```

```python
# 학습을 위해 data shape 을 변경. 연산을 위해 float 32 로 변경
X_train = X_train.reshape(60000, 784)
X_test = X_test.reshape(10000, 784)
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')
```

```python
y_train[0]
```

```python
# one-hot encoding을 통해 target label의 값을 확률로 변경
Y_train = keras.utils.to_categorical(y_train)
Y_test = keras.utils.to_categorical(y_test)
```

```python
# 정답이 5인 경우
Y_train[0]
```

```python
X_train.shape
```

```python
X_train[0]
```

```python
# model 생성
input_shape = (784, )
```

```python
model = keras.Sequential()
model.add(keras.layers.Dense(128, activation = 'sigmoid', input_shape=(784,)))
```

```python
model.add(keras.layers.Dense(128, activation = 'sigmoid'))
model.add(keras.layers.Dense(10, activation ='softmax'))
```

```python
model.compile(optimizer='SGD', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# 학습이 오래걸릴수도 있으니 epochs 는 처음에는 작게 1 회 또는 2회 정도. -> 빠르면 성능을 높이기 위해 20회
model.fit(X_train, Y_train, batch_size=128, epochs=50)
```
