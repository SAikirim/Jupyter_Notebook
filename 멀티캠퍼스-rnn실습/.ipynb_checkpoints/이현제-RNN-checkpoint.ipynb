```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
sk = pdr.get_data_yahoo('096770.ks') # sk
df = sk['Close']; # 종가
print(df)
df.plot() # 날짜와 그날 종가
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

s_train = scaler.fit_transform(df.values.reshape(-1, 1)) # numpy로 변환후 학습
print(s_train) 
s_df = pd.DataFrame(data=s_train, columns=['stock'])
print(s_df)
s_df.plot() # 0 ~ 1 사이 값 그래프
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
length = 5
generator = TimeseriesGenerator(s_train, s_train, length=length, batch_size=1)
print(generator[0])
print(s_train[:5])
```

```python
generator[1]
```

```python
import tensorflow as tf

model_LSTM = tf.keras.models.Sequential()
model_LSTM.add(tf.keras.layers.LSTM(5, input_shape=(5,1)))
model_LSTM.add(tf.keras.layers.Dense(1))
```

```python
model_LSTM.compile(optimizer='adam', loss='mse')
```

```python
model_LSTM.fit(generator, epochs = 10)
```

```python
import numpy as np

batch = np.array(s_df[-5:].index)
batch
```

```python
forecast_LSTM = []
c_batch = batch.reshape((1, 5, 1))
print(c_batch)

for i in range(5):
  c_predict = model_LSTM.predict(c_batch)[0]
  forecast_LSTM.append(c_predict)
  c_batch = np.append(c_batch[:,1:,:], [[c_predict]], axis = 1)
```

```python
forecast_LSTM = scaler.inverse_transform(forecast_LSTM)
```

```python
print(forecast_LSTM, len(forecast_LSTM))
```

```python
import matplotlib.pyplot as plt
plt.plot(df)

print(df.index[-5:])
plt.plot(df.index[-5:], forecast_LSTM)
```
