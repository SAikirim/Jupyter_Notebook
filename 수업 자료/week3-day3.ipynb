```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import random
```

```python
from tensorflow.compat.v1 import ConfigProto, InteractiveSession
config = ConfigProto()
config.gpu_options.per_process_gpu_memory_fraction = 0.3
session = InteractiveSession(config=config)
```

```python
data = [
    [0.72,0.82,-1],
    [0.91,-0.69,-1],
    [0.03,0.93,-1],
    [0.12,0.25,-1],
    [0.96,0.47,-1],
    [0.8,-0.75,-1],
    [0.46,0.98,-1],
    [0.66,0.24,-1],
    [0.72,-0.15,-1],
    [0.35,0.01,-1],
    [-0.11,0.1,1],
    [0.31,-0.96,1],
    [0.0,-0.26,1],
    [-0.43,-0.65,1],
    [0.57,-0.97,1],
    [-0.72,-0.64,1],
    [-0.25,-0.43,1],
    [-0.12,-0.9,1],
    [-0.58,0.62,1],
    [-0.77,-0.76,1]
]
```

```python
data = np.array(data)
```

```python
data.shape
```

```python
x_data = data[:, 0:2]
y_data = data[:, 2:]
```

```python
x_data
```

```python
y_data
```

```python
model=keras.Sequential()
```

```python
model.add(keras.layers.Dense(1, activation = 'tanh', input_shape=(2,)))
```

```python
optimizer = keras.optimizers.SGD(lr=0.1)
```

```python
model.compile(optimizer=optimizer , loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(x_data, y_data, batch_size=20, epochs = 3000)
```
