```python
# 한글 폰트 해결
path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)

#sub plot 추가
fig=plt.figure(figsize=(20,20))
plt.style.use('ggplot')
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)

# 서울 -> 경기
ax1.plot(df_su2ggd.index, df_su2ggd.values, marker='o', markersize=10, color = 'b')
ax1.set_xticklabels(df_su2ggd.index, rotation='vertical', size=15)

# 서울 -> 부산
ax2.plot(df_su2bs.index, df_su2bs.values, marker='o', markersize=10, color='g')
ax2.set_xticklabels(df_su2ggd.index, rotation='vertical', size=15)

# # x축 y축 값 입력
# plt.xlabel('기간')
# plt.ylabel('이동인구')
# plt.legend(labels=['서울->부산'], loc='best')
```

```python
fig = plt.figure(figsize=(20,10))
ax=fig.add_subplot(1,1,1)

ax.plot(df_su2ggd.index, df_su2ggd.values, marker='o', markersize=10, color = 'b', label = "서울 -> 경기")
ax.plot(df_su2bs.index, df_su2bs.values, marker='o', markersize=10, color='g', label =  "서울 -> 부산")

ax.set_title("서울 -> 경기도 vs 서울 -> 부산")
ax.set_xticklabels(df_su2ggd.index,rotation="vertical", size = 15)
```
