# 매직 명령어
- 파이썬 자체에서는 존재하지 않는 명령어.
    - % 기호를 붙여서 사용
    - %magic

```python
%magic
```

```python
list_1 = [i for i in range(10)]
```

```python
list_1
```

```python
list_2 = []
for i in range(10):
    list_2.append(i)
```

```python
list_2
```

```python
%time [i for i in range(1000000)]
```

```python
%timeit [i for i in range(100000)]#평균출력
```
