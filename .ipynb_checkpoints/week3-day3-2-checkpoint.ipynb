- conv 
    - 입력데이터에 대한 특징을 추출
- RELU
    - 입력정보가 0보다 크면 그냥 내보내고, 0보다 작으면 0으로 내보낸다.
- POOLING 
    - 입력정보를 최대/최소/평균 등으로 압축하여 데이터의 연산량을 줄여준다.
    - max pooling
    - min pooling
    - average pooling

```python
from PIL import Image, ImageFilter
```

```python
import matplotlib.pyplot as plt
```

```python
#!wget https://img.hankyung.com/photo/201908/03.20402175.1.jpg
```

```python
orig = Image.open("03.20402175.1.jpg")
```

```python
kerneledge = [
    1, 1, 1,
    0, 0, 0,
    1, 1, 1
]
```

```python
convol = orig.filter(ImageFilter.Kernel((3,3), kerneledge, 1,0))
```
