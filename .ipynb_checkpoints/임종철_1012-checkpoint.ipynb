```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('./detaset/보유중인_인공지능_기술_중_주력_분야_20201012155605.xlsx', header=2)
```

```python
df.info() # df.isnull()
```

```python
df_1 = df[list(df.columns)[:1] + list(df.columns)[2:11]]
df_2 = df[list(df.columns)[:1] + list(df.columns)[12:]]
df_2.rename({'사례수 (개).1':'사례수 (개)','머신러닝 (%).1':'머신러닝 (%)', '시각지능 (%).1':'시각지능 (%)', '청각지능 (%).1':'청각지능 (%)', '언어지능 (%).1':'언어지능 (%)', '지능형에이전트 (%).1': '지능형에이전트 (%)', '범용인공지능(AGI) (%).1': '범용인공지능(AGI) (%)'}, axis=1, inplace=True)
df_2
```

```python
df_1.set_index('종사자규모별(1)', inplace=True)
df_2.set_index('종사자규모별(1)', inplace=True)
```

```python
all_df1=df_1.loc['전체']
all_df2=df_2.loc['전체']
all_df1.sum()
```

```python
import matplotlib.font_manager as fm
path = 'C:\Windows\Fonts\malgun.ttf'
font_name = fm.FontProperties(fname=path, size= 50).get_name()
plt.rc('font', family = font_name)

#plt.style.use('default')

#plt.plot(df_1.index, df_1.values)
fig = plt.figure(figsize=(20,10))
#plt.title("2018년 인공지능 종사자규모(1위, 1위 + 2위)", size=20)
#plt.ylabel('퍼센트')
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)


ax1.plot(all_df1.index, all_df1.values, marker='o', label='1위')
ax2.plot(all_df2.index, all_df2.values, marker='o', label='1위 + 2위', color='b')

ax1.set_title("2018년 인공지능 종사자규모(1위)", size=15)
ax2.set_title("2018년 인공지능 종사자규모(1위 + 2위)", size=15)

ax1.legend(loc='bast')
ax2.legend(loc='bast')

ax1.set_ylabel('1위 퍼센트')
ax2.set_ylabel('1위 + 2위 퍼센트')
```
