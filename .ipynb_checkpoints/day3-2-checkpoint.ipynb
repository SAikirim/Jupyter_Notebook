#데이터의 전처리
- 누락되는 데이터 NaN
    - Not a number

```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.isnull()
```

```python
df.notnull()
```

```python
df.isnull().sum(axis=0)
```

```python
df_2=df.dropna(axis=1, thresh=600)
```

```python
mean_age=df_2['age'].mean(axis=0)
```

```python
df_2['age'].fillna(mean_age, inplace=True)
```

```python
df_2.iloc[888]
```

```python
df_2.notnull()
```

- 승선 도시별 승객수를 이용해 가장 많이타는 도시를 지정해준다.

```python
most_emb=df['embark_town'].value_counts(dropna=True).idxmax()
```

```python
df['embark_town'].fillna(most_emb, inplace=True)
```

# 유사성을 갖는 데이터
- fillna(method='ffill') : NaN 이 있는 직전행에 있는 값으로 치환
- fillna(method='bfill') : NaN 이 있는 직전행에 있는 값으로 치환

```python
df = sns.load_dataset('titanic')
```

```python
print(df['embark_town'][825:830])
```

```python
df['embark_town'].fillna(method='ffill', inplace = True)
```

```python
print(df['embark_town'][825:831])
```

```python
print(df['embark_town'][825:831])
```

```python
df['embark_town'].fillna(method='bfill', inplace = True)
```

```python
print(df['embark_town'][825:831])
```

# 중복 데이터 처리
- 중복여부를 확인 duplicated()
- df.drop_duplicates()

```python
import pandas as pd
df=pd.DataFrame({'c1':['a', 'a', 'b', 'a', 'b'],
                  'c2':[1, 1, 1, 2, 2],
                  'c3':[1, 1, 2, 2, 2]})
```

```python
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset=['c2','c3'])
```

```python
df
```

# 데이터 표준화
- 단위 환산
    - 직접 수식 입력
- 자료형 변환
    - df[""].astype()
- 정규화
    - (x-x_min)/(x_max - x_min)
- 표준화
    - x-x_mean/std(x) ~ (0,1)
