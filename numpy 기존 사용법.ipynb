# linalg 모듈
- 선형대수에 필요한 역행렬과 행렬식에 관한 함수를포함하고 있다.
    - det : 행렬식을 계산
    - inv : 역행렬 (정사각 행렬만 가능)
    - ---------------------------
    - eig : 고윳값 고유벡터 계산
    - qr : qr분해 계산
    - svd : 특잇값 분해 계산
    - matrix_rank : 행렬 의 rank를 반환한다.
    
    

```python
np.linalg.solve(w,y)
```

```python
np.linalg.matrix_rank(w) # 독립차원의 차수
```

```python
np.ndim(w) #실제 차원은 아니다. tensor기준의 차원
```
