# ndarray 생성하기

```python
np.zeros(10) # 메모리를 초기화 하고 0으로 값을 채워 넣는다.
```

```python
np.empty(10) # zeros 비슷하지만, 메모리를 초기화 하지 않고 메모리값을 반환한다.
```

```python
np.empty(30)
```

```python
np.ones((3,4)) #메모리를 초기화 하고 1로 채워 넣는다.
```

```python
np.full((3,4),10)
```

```python
# ndarray 는 권장사항 아님
np.ndarray((3,))
```
