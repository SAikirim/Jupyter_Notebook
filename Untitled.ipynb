```python
#plt.plot(data, x, y)
# 한글 폰트 해결
path = 'C:/Windows/Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=path, size= 50).get_name()
plt.rc('font', family = font_name)

#스타일 지정
plt.style.use('ggplot')
#sub plot 추가
fig = plt.figure(figsize=(20,20))  # 개체 생성
ax1=fig.add_subplot(2,1,1)
ax2=fig.add_subplot(2,1,2)


# 서울 - > 경기
ax1.plot(df_su2ggd.index, df_su2ggd.values, marker='o', markersize=10, color = 'b')
ax1.set_xticklabels(df_su2ggd.index, rotation = 'vertical', size=15)
ax1.legend(labels=['서울 경기'], loc='bast', fontsize=15)

# 서울 - > 부산
ax2.plot(df_su2bu.index, df_su2bu.values,  marker='o', markersize=10, color = 'g')
ax2.set_xticklabels(df_su2ggd.index, rotation = 'vertical', size=15)
ax2.legend(labels=['서울 부산'], fontsize=15)

plt.show()
```

```python
fig = plt.figure(figsize=(20,10))
ax=fig.add_subplot(1,1,1)

# 서울 - > 경기, 서울 - > 부산
ax.plot(df_su2ggd.index, df_su2ggd.values, marker='o', markersize=10, markerfacecolor = 'b',label='경기')
ax.plot(df_su2bu.index, df_su2bu.values,  marker='o', markersize=10, color = 'g',label='부산')

# 범례 표시
ax.legend(loc='bast') 

ax.set_title('서울 -> 경기도 VS 서울 -> 부산', size=20)
# 축 이름 추가
ax.set_xlabel('기간', size=12)
ax.set_ylabel('이동 인구수', size=12)
# 축 눈금 라벨 지정 및 75도 회전
ax.set_xticklabels(df_su2ggd.index, rotation = 75, size=13)
# 축 눈금 라벨 크기
ax.tick_params(axis='y', labelsize=13)

plt.show()
```

```python
most_emb = df['embark_town'].value_counts(dropna=True).idxmax()
```

```python
df['embark_town'].fillna(most_emb, inplace=True)
```

# 유사성을 갖는 데이터
- fillna(method='ffill') : NaN이 있는 직전행에 있는 값으로 치환
- fillna(method='bfill') : NaN이 있는 다음행에 있는 값으로 치환

```python
df = sns.load_dataset('titanic')
df['embark_town'][825:831]
```

```python
df['embark_town'].fillna(method='ffill', inplace=True)
df['embark_town'][825:831]
```

```python
df['embark_town'].fillna(method='bfill', inplace=True)
df['embark_town'][825:831]
```
